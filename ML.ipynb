```python
np.random.seed(7)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X=df.drop("AQI",axis=1)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
